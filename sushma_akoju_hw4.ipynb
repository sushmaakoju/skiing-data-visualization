```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['Vertical_feet_per_run', 'Distance_miles',
       'Max_speed_in_mph']])
```

```python
km_scores= []
km_silhouette = []
km_db_score = []
for i in range(2,df.shape[1]):
    km = KMeans(n_clusters=i, random_state=1301).fit(df_scaled)
    preds = km.predict(df_scaled)

    print(f'Score for number of cluster(s) {i}: {km.score(df_scaled):.3f}')
    km_scores.append(-km.score(df_scaled))

    silhouette = silhouette_score(df_scaled,preds)
    km_silhouette.append(silhouette)
    print(f'Silhouette score for number of cluster(s) {i}: {silhouette:.3f}')

    db = davies_bouldin_score(df_scaled,preds)
    km_db_score.append(db)
    print(f'Davies Bouldin score for number of cluster(s) {i}: {db:.3f}')

    print('-'*100)
```

```python
df_plot = pd.DataFrame({'Number of clusters':[i for i in range(2,df.shape[1])],'kmean score': km_scores})
alt.Chart(df_plot).mark_line(point=True).encode(
    x=alt.X('Number of clusters:N', axis=alt.Axis(labelAngle=0)),
    y=alt.Y('kmean score'),
    tooltip=[alt.Tooltip('kmean score', format='.2f')]
).properties(
    title='Kmean score ifo number of cluster'
).properties(
    title='The elbow method for determining number of clusters',
    width=400
).interactive()
```

```python
km = KMeans(n_clusters=3, random_state=1301).fit(df_scaled)
preds = km.predict(df_scaled)
twod_pca = PCA(n_components=2)
X_pca = twod_pca.fit_transform(df_scaled)
df_km = pd.DataFrame(data={'pca1':X_pca[:,0], 'pca2':X_pca [:,1], 'cluster':preds})

pd.Series(preds).value_counts()
```

```python

all_features = df[['Vertical_feet_per_run', 'Distance_miles',
       'Max_speed_in_mph']].columns
for i,c in enumerate(all_features):
    df_km[c] = df_scaled[:,i]

domain = [0, 1, 2]
range_ = ['red',  'darkblue', 'green']

selection = alt.selection_multi(fields=['cluster'], bind='legend')

pca = alt.Chart(df_km).mark_circle(size=100).encode(
    x='pca1',
    y='pca2',
    color=alt.Color('cluster:N', scale=alt.Scale(domain=domain, range=range_)),
    opacity=alt.condition(selection, alt.value(1), alt.value(0.1)),
    tooltip = list(all_features)
).add_selection(
    selection
)

pca.interactive()
```

```python
brush = alt.selection(type='interval')

domain = [0, 1, 2]
range_ = ['red', 'darkblue', 'green']

points = alt.Chart(df_km).mark_circle(size=60).encode(
    x='pca1',
    y='pca2',
    color = alt.condition(brush, 'cluster:N', alt.value('lightgray')),
    tooltip = list(all_features)
).add_selection(brush)

boxplots = alt.vconcat()
for measure in all_features:
    boxplot = alt.Chart(df_km).mark_boxplot().encode(
            x =alt.X(measure, axis=alt.Axis(titleX=470, titleY=0)),
    ).transform_filter(
            brush
    )
    boxplots &= boxplot

chart = alt.hconcat(points, boxplots)
# chart.save('cluster_pca_n_3.html'))
chart
```

```python
# Define the degree of the polynomial fits
degree_list = [1]

base = alt.Chart(df).mark_circle(color="black").encode(
        alt.X("sno"), alt.Y("Max_speed_in_mph")
)

polynomial_fit = [
    base.transform_regression(
        "sno", "Max_speed_in_mph", method="poly", order=order, as_=["sno", str(order)]
    )
    .mark_line()
    .transform_fold([str(order)], as_=["degree", "Max_speed_in_mph"])
    .encode(alt.Color("degree:N"))
    for order in degree_list
]

alt.layer(base, *polynomial_fit).properties( title="Regression fit Transformation  for Max Speed in mph")
```
